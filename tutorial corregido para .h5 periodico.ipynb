```python

f = h5py.File(r'/home/acastello/halos_list_con_headers.h5', 'r')


pos_hx     = df['X']    #Mpc/h
pos_hy     = df['Y']    #Mpc/h
pos_hz     = df['Z']    #Mpc/h
vel_hx     = df['VX']   #km/s
vel_hy     = df['VY']   #km/s
vel_hz     = df['VZ']   #km/s
mass_h     = df['Mvir'] #Msun/h

pos_h = np.stack([pos_hx, pos_hy, pos_hz], axis=1)
vel_h = np.stack([vel_hx, vel_hy, vel_hz], axis=1)

# Para la constante gravitacional G en unidades apropiadas, km^2 Mpc / (Msun s^2))
G = -6.67430e-11  # m^3 / (kg s^2)
G_km2_Mpc = G * (1e-3)**2 * (3.0857e22) / (1.98847e30)  # convierte a km^2 Mpc / (Msun s^2)
 

for i in range(pos_h.shape[0]):
    x0, y0, z0 = pos_h[i]
    vx0, vy0, vz0 = vel_h[i]
    m = mass_h[i]
    # Llamo a la función que quiera con esos valores

box_size = 500  # Mpc/h

def potencial(x, y, z):
    dx = x - pos_h[:, 0]
    dy = y - pos_h[:, 1]
    dz = z - pos_h[:, 2]

    # Aplicar condiciones de borde periódicas
    dx = dx - box_size * np.round(dx / box_size)
    dy = dy - box_size * np.round(dy / box_size)
    dz = dz - box_size * np.round(dz / box_size)

    r = np.sqrt(dx**2 + dy**2 + dz**2) + 1e-6  # evitar división por cero

    phi = G_km2_Mpc * np.sum(mass_h / r)
    return phi


def potencialpunto(x, y, z):
    dx = x - pos_h[:, 0]
    dy = y - pos_h[:, 1]
    dz = z - pos_h[:, 2]

    # Aplicar condiciones de borde periódicas
    dx = dx - box_size * np.round(dx / box_size)
    dy = dy - box_size * np.round(dy / box_size)
    dz = dz - box_size * np.round(dz / box_size)

    r = np.sqrt(dx**2 + dy**2 + dz**2) + 1e-6
    vr = vel_h[:, 0]*dx + vel_h[:, 1]*dy + vel_h[:, 2]*dz

    phipunto = G_km2_Mpc * np.sum(mass_h * vr / r**1.5)
    return phipunto


def potenciali(x, y, z, i):
    dx = x - pos_h[i, 0]
    dy = y - pos_h[i, 1]
    dz = z - pos_h[i, 2]

    # Aplicar condiciones de borde periódicas
    dx = dx - box_size * np.round(dx / box_size)
    dy = dy - box_size * np.round(dy / box_size)
    dz = dz - box_size * np.round(dz / box_size)

    r = np.sqrt(dx**2 + dy**2 + dz**2) + 1e-6
    phi_i = G_km2_Mpc * mass_h[i] / r
    return phi_i


def potencialpuntoi(x, y, z, i):
    dx = x - pos_h[i, 0]
    dy = y - pos_h[i, 1]
    dz = z - pos_h[i, 2]

    # Aplicar condiciones de borde periódicas
    dx = dx - box_size * np.round(dx / box_size)
    dy = dy - box_size * np.round(dy / box_size)
    dz = dz - box_size * np.round(dz / box_size)

    r = np.sqrt(dx**2 + dy**2 + dz**2) + 1e-6
    vr = vel_h[i, 0]*dx + vel_h[i, 1]*dy + vel_h[i, 2]*dz

    phipunto_i = G_km2_Mpc * mass_h[i] * vr / r**1.5
    return phipunto_i

# Para comparar:
x, y, z = 0, 0, 0
phipunto_i = potencialpuntoi(x, y, z, i=0)
phipunto_total = potencialpunto(x, y, z)
phiprint = potencial(x,y,z)
phi_i = potenciali(x,y,z, i=0)

print("Primer término individual:", phipunto_i)
print("Suma total:", phipunto_total)
print("potencial:", phiprint)
print("Primer término potencial:", phi_i)
    
```

```python
import numpy as np

# Constante física
c = 299792.458  # km/s

# Esta función ya debe estar definida antes:
# def potencialpunto(x, y, z): ...

def direccion_esferica(theta, phi):
    """
    Devuelve el vector unitario (x, y, z) a partir de (theta, phi)
    """
    return np.array([
        np.sin(theta) * np.cos(phi),
        np.sin(theta) * np.sin(phi),
        np.cos(theta)
    ])

def integrar_en_direccion_esferica(theta, phi, r_min=0, r_max=100, n_points=500):
    """
    Integra la función phipunto a lo largo de una dirección esférica (theta, phi).
    """
    direccion = direccion_esferica(theta, phi)
    r_values = np.linspace(r_min, r_max, n_points)
    phipunto_vals = []

    for r in r_values:
        x, y, z = r * direccion
        phipunto_vals.append(potencialpunto(x, y, z))  # Tu función real

    phipunto_vals = np.array(phipunto_vals)
    integral = np.trapezoid(phipunto_vals, r_values)
    return -(2 / c**3) * integral
```

```python
n_theta = 10
n_phi = 10

theta_vals = np.linspace(0, np.pi, n_theta)
phi_vals = np.linspace(0, 2*np.pi, n_phi)

I_mapa = np.zeros((n_theta, n_phi))

for i, theta in enumerate(theta_vals):
    for j, phi in enumerate(phi_vals):
        I_mapa[i, j] = integrar_en_direccion_esferica(theta, phi)
```
